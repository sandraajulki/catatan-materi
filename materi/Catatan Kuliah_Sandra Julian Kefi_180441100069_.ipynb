# 1. CLASS AND OBJECT 

Pengertian:
* class merupakan template(cetakan) yang digunakan untuk mendefinisikan objek secara umum.
* method merupakan suatu bagian dari class yang digunakan untuk mendefinisikan, menampilkan, mengolah obyek(method harus memiliki parameter self agar bisa dipanggil).
* atribut merupakan karakteristik yang melekat pada suatu objek(biasanya didefinisikan sebagai variabel).
* Objek dibuat menggunakan konstruktor kelas. Objek ini kemudian akan disebut turunan dari kelas.
* Instance dari class. Jika class secara umum merepresentasikan (template) sebuah object, maka sebuah instance adalah representasi nyata dari class itu sendiri.

# 2. INHERITANCE AND TYPE


* Inheritance : pewarisan atau turunan.
* Inheritance adalah konsep pemrograman dimana sebuah class (induk) dapat ‘menurunkan’ property dan method yang dimilikinya kepada class lain(anak). Class yang akan ‘diturunkan’ bisa disebut sebagai class induk (parent class), super class, atau base class. Sedangkan class yang ‘menerima penurunan’ bisa disebut sebagai class anak (child class), sub class.

Type dari Inheritance :
* a. Pass adalah kondisi dimana suatu method ini diberikan pass karena sama dengan method induk , dan methodnya bisa ditulis ataupun tidak karena tidak berpengaruh.
* b. Overriding adalah kondisi dimana suatu method yang berasal dari class induk ingin anda modifikasi atau ditambahkan beberapa fitur sesuai kebutuhan pada class anak, disinilah peran dari 'overriding method'.
* c. super() adalah mengembalikan atribut dan metode dari super objek (induk) suatu kelas.(mengambil atribut dari induk).
* d. Multiple Inheritance adalah kondisi dimana kita dapat mewariskan dari beberapa level class, superclass dengan mewariskan ke subclass kemudian subclass mewariskan ke class lain.

# 3. POLYMORPHISME

* OOP (object orented programming) memilki 4 bagian yaitu:
1. inheritance
2. polymorphisme
3. Abstrak
4. Encapsolation
keuntungan dari OOP (object orented programming) adalah kita bisa memanggil kodenya berkali kali basic darri OOP adalah class dan object
* class merupakan template(cetakan) yang digunakan untuk mendefinisikan objek secara umum.
* Objek merupakan turunan dari kelas.
* Inheritance adalah konsep pemrograman dimana sebuah class (induk) dapat ‘menurunkan’ property dan method yang dimilikinya kepada class lain(anak).
keuntungan inheritance dari class lain adalah menggunakan kembali(reuse), mengembangkan(extend), memodifikasi(modify).


* Poly : many (banyak)
* Morphism : form (bentuk)
* Polymorphism merupakan kemampuan beberapa obyek yang masing-masing berbeda kelas (namun masih terikat inheritance) dalam merespon permintaan kepada suatu method yang sama melalui cara pemanggilan yang spesifik.
* polymorphism dengan fuction adalah membuat fungsi yang dapat memanggil objek apapun.
* polymorphism dengan metode adalah menggunakan method tanpa peduli dari class mana objek ada.
* class abstrak adalah class ini tidak memiliki implementasi tetapi mendefinisikan struktur (dalam bentuk fungsi) yang harus dimiliki semua bentuk.
